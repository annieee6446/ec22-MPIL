## Purpose
Magnetic polarity inversion lines (MPILs) detected from solar active regions with features engineered from them are recognized as one of the essential keys for predicting the characteristics of explosive and eruptive instabilities such as solar flares and coronal mass ejections (CME). We have built a software framework that can detect MPILs from the line-of-sight (LoS) or the radial component of the magnetic field vector in solar active region magnetogram patches. As part of our efforts, we applied this detection method on magnetogram patches obtained from HMI Active Region Patches (HARP) data product.


## Technical contributions
- We share our tool in the notebook and present the inner works of our detection functions as well as visualization for the MPIL detection technique over a set of eruptive active regions from Solar Cycle 24.
- We also provide a qualitative evaluation of our module and exploratory analyses that highlight the relationships between the time evolution characteristics of MPILs in addition to a video visualization tool for the MPILs and related data products.
- The tools and code implemented in the notebook have been developed with user defined thresholds that allow easy modification for analysis.


## Methodology
A brief summary of the detection method is as follows: 

First, we identify the positive and negative polarity regions by applying a magnetic field strength threshold of +/-100 Gauss. Then, we obtain the boundaries of positive and negative polarity regions by using an edge detection technique and apply intersection and thinning morphological operations along with a magnetic field strength filter to identify coarse MPILs. Lastly, we generate final MPILs by imposing a size filter that removes the relatively small disconnected components/lines and retains more significant ones. 

In addition to the MPIL detection tool, we provide feature extraction functions to obtain the properties of MPILs (i.e., MPIL size, the area of polarity inversion, the masked unsigned flux of enclosing MPIL, convexity, eigenvalues, fractal dimension, and Hu moments) and MPIL-related binary masks of rasters. These vector and raster data components are structured as time series and will be shared as datasets. 


## Results 
This work provides a detection and visualization analysis tool for identifying MPILs and it can also be parameterized based on user's input threshold. We present detection results for the NOAA AR 11158 (HARP 377) CEA magnetogram series. Our detection results show that our approach can successfully identify PILs and visualize the identified PILs overlay on magnetograms in both images and videos.

# Parameter definitions
The tools and code implemented in the notebook have been developed with user defined thresholds that allow easy modification for analysis.
- POS_GAUSS: Magnetic field strength threshold for identifying positive polarity regions
- NEG_GAUSS: Magnetic field strength threshold for identifying negative polarity regions
- EDGE_KERNEL: Size of buffered kernel in the edge detection
- MAG_FIELD_FILTER: Flux filter threshold for keeping the unsigned magnetic flux contained
- SIZE_FILTER: CEA pixel filter threshold for removing relatively small inversion lines (i.e., each CEA pixel roughly covers approximately 131,400 km^2). The MPIL size is defined as the number of contiguous pixels contained in each disconnected MPIL.

```python
POS_GAUSS = 100
NEG_GAUSS = -100
EDGE_KERNEL = 4
MAG_FIELD_FILTER = 0.95
SIZE_FILTER = 14
```

# Data import
Our PIL detection framework applies to longitudinal magnetograms. Throughout the development lifecycle of our framework, we made use of the HMI Active Region Patches (HARP) data product [1]. HARPs are essentially tracked active regions and come with both vector and LoS component of magnetic field. The LoS magnetic field product is stored in FITS format. We primarily used the definitive series with 720 seconds cadence which are mapped the Lambert Cylindrical Equal Area (CEA) projection, where each pixel roughly occupies the same physical area on the Sun. For brevity, in the samples shown below, we adopt a one hour cadence instead of the 720 seconds. 

```python
# Count the number of samples
data = glb.glob('../Sample_Input/377/*.fits')
print('Total number of sample images: ' + str(len(data)))
```

### Plot out an example of imported data

```python
# Specify the index of desire sample
data_index = 0
```

```python
hmi_magmap = sunpy.map.Map(data[data_index])

fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.show()
```

# Detection process and analysis

### Identify Polarity Region by Strength threshold
The detection process starts by identifying the positive and negative polarity regions. We generated and applied two binary masks of the opposite polarity regions onto the original maps with a magnetic field strength thresholds of user defined Gauss values. The predefined thresholds are +/- 100 Gauss.

```python
pos_377, neg_377 = dt.identify_pos_neg_region(hmi_magmap, pos_gauss = POS_GAUSS, neg_gauss= NEG_GAUSS)

masked_pos = dt.mask_img(pos_377)
masked_neg = dt.mask_img(neg_377)
```

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.imshow(masked_pos, 'cool', interpolation='none', alpha=0.2)
plt.imshow(masked_neg, 'autumn', interpolation='none', alpha=0.2)
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.show()
```

### Apply Canny Edge Detection
In this step, we obtain the boundaries of positive and negative polarity regions separately by using the Canny edge operator [2], which applies a Gaussian filter to smooth the magnetogram images and detects the overall contour of the polarity regions.

```python
pos_edge_377 = dt.edge_detection(pos_377)
neg_edge_377 = dt.edge_detection(neg_377)

masked_pos_edge = dt.mask_img(pos_edge_377)
masked_neg_edge = dt.mask_img(neg_edge_377)
```

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.imshow(masked_pos_edge, 'cool', interpolation='none', alpha=1)
plt.imshow(masked_neg_edge, 'autumn', interpolation='none', alpha=1)
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.show()
```

### Extend the thickness of edge of Polarity region
Considering the region of polarity inversion (RoPI) as when the positive and negative polarity regions are adjacent, we determine the intersection regions between spatially buffered edges of polarity regions as the RoPIs. In this part, we spatially buffer the edges of negative and positive polarity regions in binary images with a size-4 kernel to represent the expanded areas of a given polarity inversion line. 

```python
pos_dil_edge_377 = dt.buff_edge(pos_edge_377, size = EDGE_KERNEL)
neg_dil_edge_377 = dt.buff_edge(neg_edge_377, size = EDGE_KERNEL)

masked_pos_dil_edge = dt.mask_img(pos_dil_edge_377)
masked_neg_dil_edge = dt.mask_img(neg_dil_edge_377)
```

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.imshow(masked_pos_dil_edge, 'cool', interpolation='none', alpha=0.5)
plt.imshow(masked_neg_dil_edge, 'autumn', interpolation='none', alpha=0.5)
plt.show()
```

### Candidate RoPIs
The candidate RoPIs are generated based on the intersection areas of spatially buffered opposite polarity regions.

In this step, we also filter magnetogram that are potentially affected by severe projection effect (greater than 70 degrees East-West) using SPEI index and patch corner checking.

```python
pil_orig, label_orig = pil_dt.PIL_detect(pos_gauss = POS_GAUSS, neg_gauss= NEG_GAUSS, size_kernel = EDGE_KERNEL)
```

```python
data_num = pil_dt.check_header(hmi_magmap)
```

```python
masked_pil = dt.mask_pil(label_orig[data_num])
```

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.imshow(masked_pil, 'spring', interpolation='none', alpha=1)
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.show()
```

### Candidate MPILs
#### Remove part based on magnetic field strength filter threshold
To generate the candidate PILs, we utilize an unsigned magnetic field strength filter on the areas covered by disconnected RoPIs, which ensures that the selected candidate PILs represent within the user defined threshold of the unsigned magnetic flux contained in RoPIs and removes the candidate PILs with relatively low flux buildup around them.

```python
strength_label = pil_dt.filter_by_strength(threshold = MAG_FIELD_FILTER)
```

```python
masked_filter = dt.mask_pil(strength_label[data_num])
```

#### Light yellow parts are removed based on the filter threshold

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.imshow(masked_pil, 'Wistia', interpolation='none', alpha=1)
plt.imshow(masked_filter, 'spring', interpolation='none', alpha=0.8)
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.show()
```

### Morphological thinning operations
Followed by applying a morphological thinning operation [3] to the candidate RoPIs, we eliminate the majority of the foreground pixels and preserve the topology of the primary shapes for the MPILs.

```python
thin_df, thin_binary = pil_dt.thin_strength_label(strength_label)
```

```python
masked_thin = dt.mask_pil(thin_binary[data_num])
```

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.imshow(masked_thin, 'spring', interpolation='none', alpha=1)
plt.xlabel('Carrington Longitude [deg]',fontsize = 12)
plt.ylabel('Latitude [deg]',fontsize = 12)
plt.show()
```

### Further Filter PIL based on Length/Size threshold
We apply a size filter of user defined CEA pixels (i.e., each pixel roughly covers approximately 131,400 km^2) to remove relatively small MPILs and retain cleaner and more significant ones. The predefined size threshold is 14 by 14 CEA pixels.

```python
filter_image = pil_dt.filter_by_size(thin_df, thin_binary, size_threshold = SIZE_FILTER)
```

```python
masked_size_filter = dt.mask_pil(filter_image[data_num])
```

#### Light yellow parts are removed based on the filter

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.imshow(masked_thin, 'Wistia', interpolation='none', alpha=1)
plt.imshow(masked_size_filter, 'spring', interpolation='none', alpha=1)
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.show()
```

### Generate Regions of Polarity Inversion (RoPI)
To generate the Regions of Polarity Inversion (RoPI) Area, we count the number of pixels covered by the intersection of opposite polarity inversion regions.

```python
ropi = pil_dt.RoPI(thin_df)
```

```python
masked_ropi = dt.mask_pil(ropi[data_num])
```

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.imshow(masked_ropi, 'summer', interpolation='none', alpha=1)
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.show()
```

### Generate convex hull of PIL
To generate the convex hull, we count the convex closure of the set in detected MPIL.

```python
convex_pil = pil_dt.convex_pil(filter_image)
```

```python
masked_convex = dt.mask_pil(convex_pil[data_num])
```

```python
fig = plt.figure(figsize=(8,6))
hmi_magmap.plot()
plt.imshow(masked_convex, 'autumn', interpolation='none', alpha=0.2)
plt.xlabel('Carrington Longitude [deg]',fontsize = 16)
plt.ylabel('Latitude [deg]',fontsize = 16)
plt.show()
```
